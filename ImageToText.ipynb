##---------------------------------------------- OCR using OpenCV -------

```python
##---------------------------------------------- OCR using OpenCV -------------------------------------

import cv2

# preprocessing
# gray scale
def gray(img):

    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite(r"./content/drive/MyDrive/OCR/preprocess/img_gray.png",img)
    return img

# blur
def blur(img) :
    img_blur = cv2.GaussianBlur(img,(5,5),0)
    cv2.imwrite(r"./content/drive/MyDrive/OCR/preprocess/img_blur.png",img)
    return img_blur

# threshold
def threshold(img):
    #pixels with value below 100 are turned black (0) and those with higher value are turned white (255)
    img = cv2.threshold(img, 100, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY)[1]
    cv2.imwrite(r"./content/drive/MyDrive/OCR/preprocess/img_threshold.png",img)
    return img# preprocessing
```

```python

from PIL import Image
image = Image.open('./drive/MyDrive/OCR/test_img.jpg')

# summarize some details about the image
print(image.format)
print(image.size)
print(image.mode)
```

```python
import numpy as np
im = np.array(image)

# Finding contours
im_gray = gray(im)
im_blur = blur(im_gray)
im_thresh = threshold(im_blur)

contours, _ = cv2.findContours(im_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
```

```python
from google.colab.patches import cv2_imshow

# text detection
def contours_text(orig, img, contours):
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)

        # Drawing a rectangle on copied image
        rect = cv2.rectangle(orig, (x, y), (x + w, y + h), (0, 255, 255), 2)

        cv2_imshow(rect)
        cv2.waitKey()

        # Cropping the text block for giving input to OCR
        cropped = orig[y:y + h, x:x + w]

        # Apply OCR on the cropped image
        config = ('-l eng --oem 1 --psm 3')
        text = pytesseract.image_to_string(cropped, config=config)

        print(text)

contours_text(im, im_thresh, contours)
```
